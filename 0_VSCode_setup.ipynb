### Pip ou Conda ? Les managers de paquets

Les deux sont des *managers* de paquets, qui vont permettre d'installer des modules dans l'environnement virtuel. Je choisis PiP, aussi parce que c'est une méthode connue sous Linux.
On commence par installer PiP. C'est la ligne de code Python d'après.

Lorsque j'intalle PiP dans le terminal, VS Code me suggère la commande ``` & d:/ADL/23_Udemy/Virutual_env/Scripts/python.exe -m pip install -U flake8``` pour installer le module de *linting* flake8. en 
On commence par mettre à jour pip. Dans la console, la syntaxe est 
```shell
python.exe -m pip install --upgrade pip
```
Mais on peut également utiliser la commande **!** directement dans le jupyter Notebook, comme suit:
```python
!pip install --upgrade pip
```
Même si cette syntaxe n'est pas optimale si on a plusieurs environnements virtuels. On va préférer la syntaxe qui installe les paquets et donc les modules dans l'environnement virtuel sélectionné. Note le module **sys**, obligatoire à lancer une fois pour faire tourner la commande.



````python
import sys
!{sys.executable} -m pip install --upgrade pip
````


### Packages à installer dans l'environnement virtuel
... Ainsi que listés par feu Mike lui-même. Les modules installent eux-même leurs dépendances. On finit par une liste des modules pour être bien sûr de la palanquée de modules qu'on a récupéré.

On se fie à la [documentation de VS-code Jupyter](https://github.com/microsoft/vscode-jupyter/wiki/Installing-Python-packages-in-Jupyter-Notebooks) pour la syntaxe des commandes.
On préfèrera donc la commande ```%``` à celle de ```!``` pour installer des modules dans un [environnment virtuel sous VS Code](https://packaging.python.org/en/latest/guides/installing-using-pip-and-virtual-environments/#creating-a-virtual-environment).
On se fie à la syntaxe de ce [dernier lien](https://jakevdp.github.io/blog/2017/12/05/installing-python-packages-from-jupyter/)

````python
import sys  # si ce n'est pas déjà fait
!{sys.executable} -m pip install geopandas
!{sys.executable} -m pip install matplotlib
# !{sys.executable} -m pip install psycopg2 # se connecter à du postgres # SQLAlchemy
!{sys.executable} -m pip install notebook
!{sys.executable} -m pip install PySAL # useful for statistics, usually(formerly?) a bitch of a package
````

On peut demander les mêmes commandes *via* cette syntaxe que lorsqu'on la tape dans la console.

````python
!{sys.executable} -m pip list
````

La commande en haut est donc à taper dans une ligne de commande powershell, mais on peut tout aussi bien la taper dans un Jupyter Notebook avec la commande ci-après qui ne permet néanmoins pas de contrôler dans quel environnement on fait tourner la requête.

La commande pour avoir de l'aide:

````python
!{sys.executable} -m pip -h
````
